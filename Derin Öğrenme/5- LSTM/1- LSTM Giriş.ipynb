- RNN'ler uzun dizi problemleri olduğunda teorik olarak taşıdığını düşündüğümüz geçmiş bilgiyi taşıma kabiliyetlerini kaybetmektedirler. Bunun sebebi de "Kaybolan Gradyan" problemidir. Dolayısıyla çok derin ileri yayılımlı ağlarda görüldüğü gibi katman eklendikçe ağın eğitilemez olması durumu söz konusu olmaktadır. Bu durumda LSTM yöntemi kullanılmaktadır.

- LSTM, birçok zaman adımının taşınmasını sağlayan taşıyıcı bir şerit ekleyerek geçmişteki bilgiyi geleceklerdeki periyotlara da taşıma imkanı sağlayarak dizideki bir bilgi herhangi bir noktada taşıyıcı kayışa geçirilebilir. Böylece bir sonraki zaman adımlarına ulaştırılabilir ve ihtiyaç olduğunda tekrar kullanılabilir.

- LSTM; geçmişten yani Long'dan hatırlar, yakından yani Short'tan tahmin eder.

1) Forget Gate

- LSTM'in neyi unutup neyi unutmayacağını hatırlayacağını ifade eder. 

- Bu bölümde kullanılan sigmoid fonksiyonu sayesinde girdi (x(t)), geçmiş durum bilgisi (a(t-1)) ve bias ile oluşturulmuş olan değer 0-1 arasında dönüştürüleceğinden dolayı (sigmoid fonksiyonuyla) unutma unutmama durumu için bir katkı sağlar. Değerin 0'a yakın olması unutulması ile ilişkilendirilecek.



______________________________________________________________________________________________________________________________________

2) Update (Input) Gate

- Hücre durumuna yeni bilgi eklenip eklenmeyeceğini kontrol eden kapıdır. Bu kapı, hangi yeni bilgilerin hücre durumuna ekleneceğine karar verir.

    - Giriş kapısı, girdi (x(t)) ve önceki gizli durumu (a(t-1)) kullanarak bir sigmoid fonksiyonu ile 0-1 arasında bir değer üretir.
    - Ardından, yeni aday değerler (genellikle tanh aktivasyonu ile) oluşturulur.
    - Giriş kapısı çıktısı ile aday değerler çarpılır ve hücre durumuna eklenir.

- Bu sayede LSTM, hangi yeni bilgilerin hafızaya alınacağını belirler ve gereksiz bilgilerin eklenmesini engeller.
- Kısaca: Update (Input) Gate, hücreye hangi yeni bilginin ekleneceğini kontrol eder.. 



_______________________________________________________________________________________________________________________________________

3) Memory Update

- Hücre durumunun (cell state) güncellenmesini sağlar. Bu adımda, unutma kapısı (forget gate) ile eski bilgilerin bir kısmı silinir, giriş kapısı (input gate) ile yeni bilgiler eklenir. Yani, hücre durumunun yeni hali şöyle hesaplanır: 
"
yeni_hücre_durumu = (eski_hücre_durumu * unutma_kapısı_çıktısı) + (giriş_kapısı_çıktısı * aday_değerler)

- Bu sayede LSTM, geçmişten gelen önemli bilgileri tutarken, gereksiz olanları unutur ve yeni bilgileri hafızasına ekler. Bu mekanizma, uzun süreli bağımlılıkların öğrenilmesini kolaylaştırır.

- Memroy Update = Forget Gate + Update (Input Gate)



_______________________________________________________________________________________________________________________________________

4) Output Gate

- Hücrede tutulan bilginin ne kadarının bir sonraki adımda çıktı olarak verileceğini belirler.

- Output gate, hücre durumunu (cell state) ve önceki gizli durumu (hidden state) kullanarak bir sigmoid fonksiyonu ile 0-1 arasında bir değer üretir. Ardından, hücre durumunun aktivasyonu (genellikle tanh ile) ile çarpılarak, bir sonraki zaman adımına aktarılacak çıktı (hidden state) elde edilir.

- Kısaca:
    * Output gate, hücredeki bilginin ne kadarının çıktı olarak kullanılacağını kontrol eder.
    * LSTM'nin hem çıktısını hem de bir sonraki gizli durumunu belirler.

* çıktı_kapısı = sigmoid(W_o * [a(t-1), x(t)] + b_o)
<br>a(t) = çıktı_kapısı * tanh(hücre_durumu)
