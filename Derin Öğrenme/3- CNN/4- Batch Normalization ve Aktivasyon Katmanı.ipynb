# Batch Normalization Katmanı

- Zorunlu bir katman değildir. 

- Normalizasyon

    - Normalleştirmek ve standartlaştırmak için kullanılan bir ön işleme tekniğidir.

    - Eğitim sürecini hızlandırır.

    - Hata fonksiyonunu optimize etmede kullanılır.

<br></br>
- Batch Normalization
    - Bir sinir ağının gizli katmanları arasında yapılan bir normalleştirme işlemidir.

    - Tüm veriyi değil de mini gruplar halinde normalleştirme işlemi yapılmaktadır.

    - Normalleştirme işlemi "mini batch ortlama" ve "mini batch varyans" üzerinden gerçekleştirilmektedir.
