# Bilgisayar Gözünden Görüntü






# Evrişimli Sinir Ağları (CNN - Convolutional Neural Network)



- Temel olarak görütüleri sınıflandırmak ve nesne tanıma yapmak için kullanılır.

- Bir girdi görüntüsü alıp görüntüdeki çeşitli görünüşleri/nesneleri birbirinden ayırabilen derin öğrenme ağ çeşididir.

- <h5 style="color:red"><b>Evrişimsel yapıda olması, görüntüdeki örüntüyü basitten karmaşığa doğru analiz edebilmesini ifade etmektedir. Ve aynı zamanda eğitim aşamasında öğrenilmesi gereken parametre sayısını da azaltabilmeltedir.</b></h5>

- Evrişimli Sinir Ağları, 1996 yılında **Yann LeCun (Facebook Ai Lab Direktörü)** tarafından bir makale yayınlanmasıyla ortaya çıkmıştır.

-  **Bu yapının dünyaca duyulmasını sağlayan en önemli özelliği,** 2012 yılında düzenlenen "ImageNet" isimli yarışmadır. Bu yarışma, büyük ölçekli görsel tanıma yarışmasıdır. Yüzlerce obje kategorisi ve milyonlarca resimden obje sınıflandırmaya yönelik olarak yapılan bir yarışmadır. 

- Örneğin günümüzdeki sürücüsüz araçların yoldan geçen yayayı, trafik ışıklarını tanımlayabilmesinin temelinde CNN yatmaktadır.

- Aşağıda CNN yapısının ilk test ediliş görseli yer almaktadır. 
    - 1989 yılında Bell Labs'ta Yann LeCun tarafından CNN ile Geriye Yayılım Algoritması birleştirilerek el yazısıyla yazılmış rakamların tanınması üzerine bir uygulama gerçekleştirmiştir.

    - Yann LeCun, bu ağın ismini **LeNet** olarak vermiştir. 

    - Yapılan bu çalışma ABD posta servisi tarafından posta kodlarının zarflardan okunması işleminin otomatikleştirilmesi için kullanılmıştır.
