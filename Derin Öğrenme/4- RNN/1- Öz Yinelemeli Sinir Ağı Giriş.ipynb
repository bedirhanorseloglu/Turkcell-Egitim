- RNN, "dizi" şeklindeki veriler (zaman serisi verileri, doğal dil işlemeleri) için çalışmak üzere tasarlanmış özel bir yapay sinir ağı türüdür.
    - zaman serisi verileri -> bir hisse senedinin yarınki fiyatını tahmin etmek istediğimizde bugün ve bugünden önceki değerleri göz önünde bulundurmamız gerekir. Çünkü gelecekle ilgili yapılacak tahminler geçmiş bilgilerden etkilenmektedir.

    - doğal dil işlemeleri -> bir cümle kendisinden önceki cümlelerden etkileniyor olacaktır. Bir cümle içindeki bir kelime de önceki kelimeden etkileniyor olacaktır. Yani aralarında bir "dizi" ilişki vardır.

- Klasik ileri beslemeli yapay sinir ağları modelleri birbirinden bağımsız veri noktaları içerirken RNN, birbirine bağlı olan veri noktaları içerir.

- Peki RNN'de bu dizi şeklinde çalışma imkanını veren şey nedir?
    - "Bellek" kavramı -> dizinin bir sonraki çıktısını oluşturmak için önceki girdilerin durumlarını (bilgilerini) saklamaya yardımcı olur.

- RNN'in çalışma mantığı:
    - a(0) kendinden önce bir durum olmadığı için 0'lardan oluşan bir matristir. Durum'umuz ağrılıkla çarpılır, input da ağırlıkla çarpılır. Bias değeri eklenir. Daha sonra aktivasyon fonksiyonundan geçerek bir output oluşturur. Bir durumun outputu kendinden sonrakine durum olarak gönderilir. Output ağırlıkla çarpılarak tahmin edilen değer elde edilir. Bu tahmin edilen değer optimizasyonda kullanılır.
    - <b>a ifadeleri yani durumlar geçmişten gelen bilgilerin tutulduğu yerdir.
