- Burada rastgele ağırlıklar kullanılarak aktivasyon fonksiyonuyla işleme soktuktan sonra  y yani tahmini değerlerimizi elde ettiğimizi söylemiştik. Tahmini değerlerimiz olan y değerleri gerçek değerlerle karşılaştırıldığında bir Loss değeri elde edilir. Bu her adımda bu Loss değeri elde edilir ve toplanılarak toplam Loss değeri elde edilir. Daha sonra bu Loss değerini düşürmek için geri yayılım ile birlikte ağırlıkları güncelleyeceğiz.
