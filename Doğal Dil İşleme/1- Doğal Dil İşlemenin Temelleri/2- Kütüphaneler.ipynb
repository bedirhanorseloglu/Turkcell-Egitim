### NLTK (Natural Language Toolkit):
- Tokenizasyon, post etiketleme gibi pek çok alanda kullanılır.

### spaCy:
- Modern ve hızlı bir NLP kütüphanesidir.
- Önceden eğitilmiş modellerle entegre araçlar sunar.
- İsim tanıma, özetleme, bağlantı çıkarımı gibi farklı algoritmalar için kullanılabilir.

### Hugging Face:
- BERT, GPT gibi büyük dil modellerini kullanmamızı sağlayan bir kütüphanedir.
- Önceden eğitilmiş LLM (Large Language Model) dil modellerini ve transfer öğrenimi için uygundur.

### TextBlob:
- Python için basit bir NLP kütüphanesidir.
- Temel dil işleme görevlerini hızlıca gerçekleştirir.
- Örneğin; duygu analizi, metin sınıflandırma gibi

### Gensim:
- Metin madenciliği ve word embedding için kullanılır.
