- Transformers, doğal dil işleme ve diğer yapay zeka alanlarında son yıllarda devrim niteliğinde yenilikler getiren bir mimaridir.

- İlk olarak 2017 yılında Google tarafından yayımlanan "Atention is All You Need" adlı makalede tanıtılmıştır.

- Neden transformers:
    - Bağlamı daha iyi anlama (uzun mesafeli ilişkileri yakalamada oldukça başarıldır)
    - Paralel işleme yeteneği (paralel işleme yapabildiği için eğitim ve test süreçlerinde daha hızlıdır. Derin öğrenme modelleri sıralı işlem yaparken transformers paralel işlem yapabiliyor)
    - Çeşitli NLP görevlerinde kullanım (dil modelleme, metin sınıflandırma, dil üretimi, çeviri, özetleme...)
    - Önceden eğitilmiş modellerin yeniden kullanımı (transformers modelleri genellikle büyük veri setlerini kullanarak önceden eğitilirler. Bu modeller farklı görevler için fine tuning (ince ayar) yapılarak kullanılabilirler)

- En bilinen transformers modelleri:
    - BERT (Bidirectional Encoder Representations from Transformers), bir kelimenin anlamını betimlerken o kelimenin hem önündeki hem arkasındaki kelimeleri dikkate alır. Özellikle metin sınıflandırma, adlandırılmış varlık tanıma, soru yanıtlama gibi görevlerde sıklıkla kullanılır.
    - GPT (Generative Pre-trained Transformer), dil modeli olarak geliştirilmiştir. Özellikle dil oluşturma gibi görevlerde kullanılır. Verilen bir başlangıç metninden bu metinin devamını kendisi türetebilir. Tek yönlüdür (directional). Yani metini sağa ya da sola doğru işler. Otoagresiv bir modeldir. Yani bir kelimeyi tahmin ederken önceki kelimelere dayanır.

- Attention: modelin belirli girdi parçalarına farklı derecelerde dikkat göstermesine olanak tanır. Özellikle bir kelimenin diğer kelimelerle olan ilişkisini anlamak için kullanılır. 

- Input Embedding: girdi verilerini modelin işleyebileceği bir formata dönüştürmek için kullanılan bir tekniktir (Metin temsili yöntemlerimiz).

- Multi-Head Attention: attention mekanizmasının birden fazla başlıkla (head) çalıştığı bir tekniktir. Bir cümledeki her kelime, diğer kelimelerle olan ilişkilerini farklı açılardan öğrenmek isteyebilir. Örneğin "kedi" kelimesinin "hızlıdır" kelimesiyle olan ilişkisini anlamak için birden fazla dikkat başlığı kullanılabilir:
    - Başlık 1-> "kedi" ve "hızlıdır" arasındaki anlam ilişkisini öğrenir
    - Başlık 2-> "kedi" ve "hızlıdır" arasındaki dil bilgisi ilişkisini 
    - Başlık 3-> "kedi" kelimesinin cümledeki konumunu öğrenir

- Masked Multi-Head Attention: modelin gelecekteki kelimeleri görmesini engeller. Yani model sadece geçmiş bilgileri kullanarak tahminde bulunur. Örneğin "Kedi ____ hızlıdır" cümlesinde "kedi" ve "hızlıdr" arasındaki ilişkilere dayanarak boşluğa "oldukça" kelimesini tahmin eder. 

- Add & Norm: bir katman çıktı ile girdi arasındaki kısa yolu (residual connection) ekleyip ardından layer normalization uygulayan bir adımdır.
