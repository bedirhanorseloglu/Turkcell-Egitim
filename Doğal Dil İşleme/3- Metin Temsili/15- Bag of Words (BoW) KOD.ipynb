```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Örnek metinleri tanımlayalım
documents = [
    "Kedi evde",
    "Kedi bahçede"
]
```

```python
# Kütüphanemizi kullanmak için nesne oluşturuyoruz

vectorize = CountVectorizer()
```

```python
# Metinlerimizi sayısal verilere dönüştürüyoruz (vektörleştiriyoruz)

x = vectorize.fit_transform(documents)
x
```

```python
# Kelime kümemizi görelim (kedi, evde, bahçede)

print(f"Kelime Kümemiz: {vectorize.get_feature_names_out()}")
```

```python
# Vektör temsilimize bakalım

print(f"Vektör Temsilimiz: {x.toarray()}")

# İlk cümle "Kedi evde" temsili [0 1 1] -> bahçede 0, evde 1, kedi 1
# İkinci cümle "Kedi bahçede" temsili [1 0 1] -> bahçede 1, evde 0, kedi 1
# Kelime kümesindeki kelimelerin sırasına dikkat edelim: [bahçede, evde, kedi]
```
