- TF-IDF (Term Frequency - Inverse Document Frequency), metin madenciliğinde ve bilgi erişiminde sıkça kullanılan bir özellik çıkarım yöntemidir.
- BoW'da bir kelimenin belge içinde ne sıklıkla geçtiğini (frekansını) hesaplayarak işlem yapmıştık. TF-IDF (TfidfVectorizer) ise kelimelerin belgeler içinde ne kadar önemli olduğunu belirlemek için kullanılır.



- TF-IDF:
    - TF-> bir kelimenin bir belgede ne kadar sık geçtiğini ölçer (frekansını). BoW kısmı ile aynı.
    - IDF-> bir kelimenin tüm belgedeki yaygınlığını ölçer. Bir kelime çok belgede geçiyorsa o kelime çok fazla bilgi sağlamaz.
        - Örneğin IMDB veri setini düşünelim. "movie" kelimesi çok fazla geçiyor. TF-IDF'e göre "movie" kelimesinin çok bir önemi olmayacaktır. Bu kelimeyi "ayırt edici" olarak görmediği için önemsemeyecektir (TF-IDF değeri düşükse ayırt edici değildir).



(Sayaç = İlgili kelime)
