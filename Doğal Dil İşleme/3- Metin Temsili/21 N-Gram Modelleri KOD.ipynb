```python
# Kütüphanelerimizi import edelim
from sklearn.feature_extraction.text import CountVectorizer # daha önce bunu BoW'da kullanmıştırk ancak parametrelerini değiştirmemiştik

# BoW'da ngram_range=(1,1) yani sadece unigram kullanıyorduk default olarak.
```

```python
# Örnek metinimizi tanımlayalım

documents = [
    "Bu bir örnek metindir",
    "Bu örnek metin doğal dil işlemeyi gösterir."
]
```

```python
# CountVectorizer'ı kullanmak için bir nesne oluşturalım
# ngram_range tuple olarak verilir. içerisindeki ilk değer min ngram, ikinci değer max ngram'ı belirtir. (1,1) sadece unigram demektir. 
# Örneğin (1,2) ise unigram ve bigram demektir.

vectorizer_unigram = CountVectorizer(ngram_range=(1,1))
vectorizer_bigram = CountVectorizer(ngram_range=(2,2))
vectorizer_trigram = CountVectorizer(ngram_range=(3,3)) 
```

```python
# Vectorizer'larımızı kullanarak metinlerimizi vektörleştirelim

x_unigram = vectorizer_unigram.fit_transform(documents)
x_bigram = vectorizer_bigram.fit_transform(documents)
x_trigram = vectorizer_trigram.fit_transform(documents)
```

```python
# Kelime kümemizi bulalım
feature_names_unigram = vectorizer_unigram.get_feature_names_out()
feature_names_bigram = vectorizer_bigram.get_feature_names_out()
feature_names_trigram = vectorizer_trigram.get_feature_names_out()

print("Unigram feature names:", feature_names_unigram , "\n")
print("Bigram feature names:", feature_names_bigram, "\n")
print("Trigram feature names:", feature_names_trigram)
```
