- Word Embeddings (Kelime Gömme), doğal dil işleme (NLP) ve makine öğreniminde kullanılan bir tekniktir.
- Kelimeleri, genellikle sürekli bir vektör uzayında anlamlı temsil edecek şekilde sayısal vektörlere dönüştürür.
- Bu temsiller, kelimeler arasındaki anlamsal ve dilbilgisel ilişkileri yakalamayı hedefler.
    - BoW'de kelimelerin frekansına bakıyorduk
    - TF-IDF'te hem frekansına hem de ayırt ediciliğine bakıyorduk



- Word Embeddings Özellikleri:
    - Anlamsal benzerlik (benzer anlamlara sahip kelimeleri benzer vektörler olarak alıyoruz)
    
    - Matematiksel işlemler (kelimeleri vektörleştirdiğimiz için kelimlerle matematiksel işlemler de yapabiliyoruz. Örneğin vektör ekleme veya çıkarma)
        - king - man + women = queen
    - Kapsamlılık (genellikle büyük veri kümesine dayanarak eğitilir. Bu yüzden dildeki çeşitli ilişkileri yakalamak açısından diğer yöntemlere göre daha iyi çalışır)

- Word Embeddings Modelleri:
    - Word2Vec-> Google tarafından geliştirilen, kelimeleri vektörlere dönüştüren ve bu vektörleri dildeki ilişkileri yakalayacak şekilde eğiten bir modeldir.
    - GloVe(Global Vectors for Word Representation)-> Stanford Üniversitesi tarafından geliştirilen, word embedding temsillerini kelime ortaklıklarını yakalayacak şekilde hesaplayan bir modeldir.
    - FastText-> Facebook tarafından geliştirilen ve word embedding temsillerini kelime alt birimlerini de dikkate alarak hesaplayan bir modeldir.

    Bu 3 model de word embedding oluşturuyor ve aslında kelimelerden vektörler elde ediyor.
