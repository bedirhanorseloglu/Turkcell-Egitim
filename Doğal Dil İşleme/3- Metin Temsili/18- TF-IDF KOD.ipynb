```python
# Kütüphanemizi içe aktaralım

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
```

```python
# Örnek belgemizi oluşturalım

documents = [
    "Kedi çok tatlı bir hayvandır",
    "Kedi ve köpekler çok tatlı hayvanlardır",
    "Arılar bal üretirler"
]
```

```python
# TF-IDF kütüphanemizi kullanmak için bir nesne oluşturalım

vectorizer = TfidfVectorizer()
```

```python
# Nesnemizi kullanarak metinlerimizi vektörleştirelim

x = vectorizer.fit_transform(documents)
```

```python
# Kelime kümemizi bulalım

feature_names = vectorizer.get_feature_names_out()
feature_names
```

```python
# Sayısallaştırdığımız metinlerimizin matrisini görelim. Vektörize ettiğimiz metinleri toarray'e dönüştürelim

print(f"TF-IDF Vektör Temsilleri: {x.toarray()}")
```

```python
# Daha güzel görebilmek için bunu bir DataFrame'e çevirelim

df = pd.DataFrame(x.toarray(), columns=feature_names)
df

# ilk satır birinci cümleyi, ikinci satır ikinci cümleyi temsil ediyor.
# Burdaki verileri incelediğimizde "çok" kelimesinin birinci cümledeki değeri "0.379303" iken ikinci cümledeki değeri "0.334712"dir.
    # Bu da demek oluyor ki "çok" kelimesi birinci cümlede daha ayırt edici bir kelimedir.
    # Şu anda elimizdeki veri az olduğu için bu değerler çok anlamlı değil.

# BoW'de frekans değerleri ile temsil ediyorduk kelimeleri. Burda ise önem dereceleri olarak temsil ediyoruz.
```

```python
# Örneğin "kedi" kelimesinin tüm cümlelerdeki TF-IDF değerlerini bulalım

kedi_tfidf = df["kedi"]
kedi_tfidf
```

```python
# "kedi" kelimesinin önemini ölçebilmek için bu değerlerin ortalamasını almamız gerekiyor
kedi_tfidf.mean()
```
