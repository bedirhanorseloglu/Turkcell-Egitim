- N-Gram, bir dil modelinde kullanılan kelime veya karakter dizisinin uzunluğunu belirten bir terimdir.
- N-Gram modelleri, metinleri n kelimelik veya n karakterlik kısımlara bölerek analiz eder.
- BoW yönteminde kullandığımız bir hiperparametredir (ngram_range).



- Buradaki:
    - Uni-Gram-> n=1 -> vectorizer_unigram = CountVectorizer(ngram_range=(1,1))
    - Bi-Gram -> n=2 -> vectorizer_bigram = CountVectorizer(ngram_range=(2,2))
    - Tri-Gram-> n=3 -> vectorizer_trigram = CountVectorizer(ngram_range=(3,3)) 


- N-Gram modellerinin kullanım alanları:
    - Metin modelleme (bir dil modeli bir sonraki kelimenin tahmini için n-1 kelimeyi kullanabilir örneğin)
    - Metin sınıflandırma
    - Metin üretimi (bir metin oluştururken olası kelime dizilerini tahmin edebilir)
    - Metin Benzerliği
