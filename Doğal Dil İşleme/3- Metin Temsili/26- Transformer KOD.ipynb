Kafa karışıklığı olmasın:
    - Biz burada transformers kullanarak bir problem çözmüyoruz. Sadece metin temsil yöntemi olarak transformers kullanıyoruz

```python
# Kütüphanelerimizi import edelim
from transformers import AutoTokenizer, AutoModel    # Hugging Face tarafından sağlanan transformer modelleri için tokenizer ve model sınıfları
import torch # Facebook tarafından geliştirilen derin öğrenme kütüphanesi
```

```python
# Model ve tokenizer'ı yükleyelim
# Model olarak BERT kullanacağız. Geçen dersimizde Word2Vec kullanmıştık.

model_name = "bert-base-uncased" # BERT'in temel versiyonu, küçük harfli
tokenizer = AutoTokenizer.from_pretrained(model_name) # Modelimizi kullanarak tokenizer edeceğiz

model = AutoModel.from_pretrained(model_name) # Önceden eğitilmiş BERT modelini yüklüyoruz
```

```python
# Metnimizi tanımlayalım

text = "Transformers are amazing for natural language processing."
```

```python
# Metinizimi tokenize edelim

inputs = tokenizer(text, return_tensors="pt") # return_tensors="pt" ifadesi, çıktının PyTorch tensörleri olarak döndürülmesini sağlar
```

```python
# Modeli kullanarak metin temsilini (embedding) elde edelim

with torch.no_grad():  # Modelin eğitilmesini engellemek için
    output = model(**inputs)  # Modeli çalıştırıyoruz

# Biz burada torch.no_grad() diyerek backpropagation'ı devre dışı bırakıyoruz yani öğrenmeyi iptal ediyoruz. Yani modelimizin ağırlıkları güncellenmeyecek... 
# ...çünkü sadece önceden eğitilmiş modeli kullanıyoruz ve modelin ağırlıklarını güncellemek istemiyoruz.
# Biz sadece ileri besleme yaparak önceden eğitilmiş modelin ağırlıklarını kullanarak metin temsilini elde etmek istiyoruz.
```

```python
# Çıkışardan ilk token'ları alalım

last_hidden_state = output.last_hidden_state
first_token_embedding = last_hidden_state[:, 0, :].numpy()
print(f"İlk token: {first_token_embedding}")

# Burada bulunan çıktılar bir sinir ağı modelinden gelen ve bir metin girişinin ilk token'ına karşılık gelen sayısal vektörlerdir.
```
