- Metin temsili, bir metni sayısal veya başka bir türede bir formatta temsil etme işlemdir.
- NLP Pipeline'ımızda metin ön işleme ile text cleaning ve pre processing kısmını hallatmiştik. Şimdi metin temsili ile feature engeneering kısmına geçiyoruz.

- Metin temsili neden yapılmalıdır?
    - Bilgisayarların anlayabilmesi (bilgisayarar metinleri değil sayıları anlayabilirler)
    - Öznitelik çıkarma
    - Model eğitimi (makine ve derin öğrenmesi modellerimiz sayısal verilerle çalışır)

- Metin temsili yöntemleri (ilkel yöntemlerden gelişmiş yöntemlere sırayla):
    - Bag of Words (BoW)
    - TF-IDF (Term Frequency - Inverse Document Frequency)
    - N-Gram Modelleri
    - Word Embeddings
    - Transformer Tabanlı Metin Temsili
