```python
# Kütüphanelerimiz import edelim
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
# Veri seitimizi yükleyelim
# Veri setini biraz temizlememiz gerekiyor

df = pd.read_csv("spam.csv" , encoding="latin-1")
df = df.rename(columns={"v1": "type", "v2": "text"})
df.drop(columns={"Unnamed: 2", "Unnamed: 3", "Unnamed: 4"}, inplace=True)
df.head()
```

```python
# Kütüphanemizi kullanmak için TF-IDF vektörleştiricimizi oluşturalım

vectorizer = TfidfVectorizer()
```

```python
# Metinlerimizi vektörleştirelim

x = vectorizer.fit_transform(df["text"])
```

```python
# Kelime kümemizi bulalım

feature_names = vectorizer.get_feature_names_out()
tfidf_score = x.mean(axis=0).A1 # Her kelimenin tüm belgelerdeki TF-IDF ortalamasını bulalım
```

```python
# Sonuçları bir dataframe olarak görelim

df = pd.DataFrame({"word": feature_names, "tfidf_score": tfidf_score})
df

# Metinlerimizi temizlemediğimiz için bu şekilde saçma kelimler çıkıyor. 
# Normalde vektörleştirmeden önce metin temizliği yapmamız gerekiyordu.
# Temizlemeden vektörleştirirsek ne olacağını görmek için bunu yaptık.
```

```python
# Bu df'i sıralayarak en önemli kelimeleri görebiliriz
df = df.sort_values(by="tfidf_score", ascending=False)
df
```
