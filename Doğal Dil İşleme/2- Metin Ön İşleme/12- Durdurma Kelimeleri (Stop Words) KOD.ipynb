```python
import nltk
from nltk.corpus import stopwords
```

```python
# NLTK kütüphanesinden stop words (durdurma kelimeleri) listesini indiriyoruz
nltk.download("stopwords")
```

```python
# İngilzce stop words listesini alalım
stop_words_eng = set(stopwords.words("english"))
stop_words_eng
```

```python
# Örnek metin
text = "This is a sample sentence, showing off the stop words filtration."
```

```python
# Örnek metindeki stop words "olmayan" kelimeleri filtreleyelim
filtered_words = [word for word in text.split() if word.lower() not in stop_words]
filtered_words
```
