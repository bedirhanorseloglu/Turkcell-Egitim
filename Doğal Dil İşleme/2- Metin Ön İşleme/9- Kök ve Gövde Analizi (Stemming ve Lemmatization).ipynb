**Steming**
- Stemming (Kök bulma), kelimelerin kök formunu yani temel anlamını bulmak için kelimenin sonundaki eklerin(suffix) çıkarılması işlemidir.

- Stemming işlemi, kelimenin anlamını tamamen doğru bir şekilde elde etmeyi amaçlaMAZ. Daha ziyade, kelimenin en basit formunu bulmaya odaklanır.
    - Örnek: "koşuyor", "koştu", "koşmak" -> koş
    - "evde", "evler", "evimiz" -> ev


**Lemmatization**
- Lemmatization (Gövdeleme), kelimeleri sözlükteki temel formlarına (lemma) dönüştürme işlemdir.

- Lemmatization, kelimenin anlamını ve dilbilgisel yapısını dikkate alarak doğru bir kök bulmaya çalışır. Stemming'de kelimenin kökü anlamsız dahi olsa onu bulmaya çalışıyorduk. Burda da kök bulmaya çalışıyoruz ancak anlamlı kök. Bu nedenle lemmatization sonrası elde edilen kelime dilbilgisel olarak anlamlı ve sözlükte yer alan bir kelime olur.
    - Örnek: "koşuyor", "koştu", "koşmak" -> koşmak
    - "evde", "evler", "evimiz" -> ev

- Stemming'de anlamsız da olsa kök, lemmatization'da anlamlı kök buluyoruz.
