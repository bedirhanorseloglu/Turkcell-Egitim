```python
# Metinlerdeki fazla boşlukları temizleme
# Birçok farklı boşluk temizleme yöntemi var.

text1 = "Hello,     World!           2035"

sp = text1.split()
print(sp)
cleaned_text = " ".join(sp)
print(cleaned_text)
```

```python
#Büyük/küçük harf dönüşümleri

text2 = "Hello, World! 2035"

print(text2.lower())
print(text2.upper())
```

```python
# Noktalama işaretlerini temizleme

import string

text3 = "Hello, World! 2035. Let's clean: this text."

# punctuation = noktalama işaretleri
# noktalama işaretlerini boşlukla değiştirme
clean_text = text3.translate(str.maketrans("", "", string.punctuation))
print(clean_text)
```

```python
# Özel karakterleri temizleme (Noktalama işaretleri de özel karakterlere dahildir)

# re = regular expression (düzenli ifade)
import re

text4 = "Hello, World! 2035. Let's clean: this text. #Python @2024$"

# A-Za-z0-9 dışındakileri subturackt (\s) et yani değiştir. Neyle değiştireceği de virgülden sonra
cleaned_text = re.sub(r"[^A-Za-z0-9\s]", "", text4)
print(cleaned_text)
```

```python
# Yazım Hatalarını Düzeltme

from textblob import TextBlob

text5 = "Helıo Wirld! 2035"

TextBlob(text5).correct()   # Bu şekilde TextBlob objesi olarak dönüyor. Stringe çevirelim

cleaned_text = str(TextBlob(text5).correct())
print(cleaned_text)
```

```python
# HTML veya URL etiketlerini temizleme

from bs4 import BeautifulSoup

text6 = " <div> Hello World! 2035 </div>"

cleaned_text = BeautifulSoup(text6, "html.parser").get_text()
print(cleaned_text)
```
