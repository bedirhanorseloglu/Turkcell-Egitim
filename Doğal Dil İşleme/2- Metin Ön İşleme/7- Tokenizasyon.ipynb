Metin Ön işleme adımlarımızdan ikincisi olan "Tokenizasyon" adımını inceleyelim

- Tokenizasyon, bir metni daha küçük parçalara ayırma işlemidir.
- Bu küçük parçalar, genellikle "token" olarak adlandırılır.
- Tokenler; kelimler, cümleler ve hatta kelimler olabilir.
- Ham metni anlamlı ve analiz edilebilir bir hale getirmemiz için önemlidir.
