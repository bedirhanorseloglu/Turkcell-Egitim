# Metin Ön İşleme

- Metin ön işleme, ham metin verilerini analiz ve işleme için daha uygun bir formata dönüştürme sürecidir.

# Metin Ön İşleme Neden Yapılmalıdır?

- Veri kalitesini artırır
- Model performansını artırır
- Veri boyutunu yönetir

# Metin Ön İşleme Adımları



- Durdurma kelimeleri, analize az katkı sağlayan kelimelerin ayrıştırılmasıdır. Örneğin "Bu bir test cümlesidir" deki "bu" ve "bir" kelimeleri durdurma kelimeleridir. "Test cümlesidir." de aynı anlamı ifade eder.
