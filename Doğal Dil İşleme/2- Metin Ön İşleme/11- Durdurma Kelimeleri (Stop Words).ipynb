- Durdurma Kelimleri, metinlerde genellikle anlamı çok az olan veya metnin analizi sırasında çok faydalı olmayan kelimelerdir.
- Bu kelimler genellikle bağlaçlar, edatlar, zamirler ve diğer dil bilgisi işlevi olan kelimelerdir.
- Metin işleme süreçlerinde bu kelimeleri kaldırmak, analizlerin doğruluğunu artırabilir ve metin üzerinde daha anlamlı sonuçlar elde edilmesine yardımcı olabilir.
    - Örnek: ve, bir, bu, ile, de, da, mi, o, çok, gibi...
    - and, the, is, in, to, of, it, that...
