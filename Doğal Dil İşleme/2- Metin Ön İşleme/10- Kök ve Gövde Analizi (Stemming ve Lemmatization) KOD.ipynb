### Stemming

```python
# Stemming işlemini gerçkeleştirecek olan fonksinumuzu import edelim
from nltk.stem import PorterStemmer
```

```python
# Bu kütüphaneyi kullanmak için nesnemizi oluşturalım
stemmer = PorterStemmer()
```

```python
# Stemming işlemi için örnek kelimeler tanımlayalım
words = ["running", "runner", "ran", "runs", "better", "go", "went"]
```

```python
stemmer.stem("running") # Örnek kullanım
```

```python
# Her bir kelime için kök bulma işlemini gerçekleştirelim
for word in words:
    print(f"Kelime: {word} --> Kök: {stemmer.stem(word)}")
```

### Lematization

```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
# Nesne oluşturalım
lemmatizer = WordNetLemmatizer()
```

```python
# Örnek kelimeler tanımlayalım
words = ["running", "runner", "ran", "runs", "better", "go", "went"]
```

```python
lemmatizer.lemmatize("running", pos='v') # Örnek kullanım
# pos='v' parametresi kelimenin türünü belirtir (v: verb-fiil, n: noun-isim, a: adjective-sıfat, r: adverb-zarf)    
```

```python
# Her bir kelime için lemmatization işlemini gerçekleştirelim
for word in words:
    print(f"Kelime: {word} --> Lemma: {lemmatizer.lemmatize(word, pos="v")}")
```
