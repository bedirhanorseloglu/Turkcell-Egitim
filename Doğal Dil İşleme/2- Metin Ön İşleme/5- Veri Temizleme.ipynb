"Metin Ön işleme adımlarımızdan ilki olan "Veri Temizleme" adımını inceleyelim"

- Veri temizleme, metin verilerini analize edilebilir hale getirmek için yapılan bir dizi işlemi içerir.

- Bu adımda, metinlerde bulunan hatalı, gereksiz, düzensiz veya model performansını olumsuz etkileyebilecek öğeler kaldırılır veya düzeltirlir.

# Veri Temizleme Adımları

1. Boşlukların Temizlenmesi



2. Büyük-Küçük Harf Dönüşümleri
    - Genellikle tüm metinler küçük harfe dönüştürülür.



3. Noktalama İşaretlerinin Kaldırılması
    - Normalde metnin anlamını anlamamıza yardımcı olur ancak analizde gürültü olarak kabul edilir ve kaldırılır.



4. Özel Karakterlerin Kaldırılması



5. Yazım Hatalarının Düzeltilmesi



6. HTML ve URL Temizleme
