```python
import nltk
```

```python
# Punkt, NLTK'nin cümle ve kelime tokenizasyonu için kullandığı bir model
nltk.download("punkt_tab")
```

```python
# Örnek bir text yazalım
text = "Selam! Nasılsınız? Bugün hava çok güzel."
```

```python
# Kelime tokenizasyonu
word_tokens = nltk.word_tokenize(text)
print("Kelime Tokenizasyonu:", word_tokens)
```

```python
# Cümle tokenizasyonu
sent_tokenize = nltk.sent_tokenize(text)
print("Cümle Tokenizasyonu:", sent_tokenize)
```
