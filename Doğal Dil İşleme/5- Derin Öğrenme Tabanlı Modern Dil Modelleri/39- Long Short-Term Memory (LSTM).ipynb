## LSTM Nedir

- LSTM, RNN türlerinden biridir.
- Uzun vadeli bağımlılıkları öğrenebilmek için özel olarak tasarlanmıştır.
- LSTM'in ana amacı:
    - Zaman bağımlı verilerde (ör: Hisse verileri) uzun dönem bağımlılıkları öğrenmek
    - Geleneksel RNN'lerin yaşadığı **vanishing gradient (kaybolan gradyan - öğrenmenin durması - ağırlıkların güncellenmemesi)** sorunu çözmek için geliştirilmiştir.



## LSTM'in RNN  Üzerindeki İyileştirmesi

- RNN, bağımlılıkları (hafızayı) ağın derinlerine doğru ulaştırmada problem yaşıyordu. Yani ağ derinleştikçe hafızada kayıp yaşanıyordu.
- Bu eksikliği gidermek yani hafızayı ağın derinliklerine doğru ulaştırmak için LSTM tasarlandı.


- RNN'lerde yaşanan sorunlar:
    - Vanishing Gradient (Kaybolan Gradyan) problemi
    - Uzun vadeli bağımlılıkları öğrenememe

- LSTM'nin Çözümü:
    - Hücre durumu bloğu eklenmesi -> her adımda bilgi taşıyan bir hücre durumunu tutar. Bu da bilgiyi uzun vadeli aktarır. Görseldeki Ct ilk hücreden son hücreye kadar ilerletilir.
    - LSTM'de 3 adet kapı bulunur (Girdi-Unutma-Çıktı) -> bu kapılar sayesinde hangi bilginin saklanacağına, güncelleneceğine veya unutulacağına karar verilir. Bu de gereksiz bilginin silinmesini ve önemli bilginin saklanmasını sağlar. 

    

## LSTM Mimarisi, Bileşenleri ve İşleyişi

- LSTM'nin bileşenleri:
    - Hücre Durumu (Cell State) -> LSTM açılımındaki Long Memory hücre durumu sayesinde sağlanır. Bilginin aktarıldığı uzun dönemli bellek görevi görür.
    - Unutma Kapısı (Forget Gate) -> Geçmişten gelen hangi bilginin unutulacağını veya korunacağını belirler. Gereksiz bilgileri hücre durumundan temizler.
    - Giriş Kapısı (Input Gate) -> Yeni bilginin hücre durumuna eklenmesine karar verir. Girdi bilgisinin ne kadarının hücreye aktarılacağını belirler. 
    - Çıktı Kapısı (Output Gate) ->  O anki hücre durumuna bakarak neyin çıktı olarak verileceğini belirler. 

- LSTM'nin İşleyisi:
    - Unutma Kapısı -> Giriş Kapısı -> Hücre Durumu Güncellemesi -> Çıktı Kapısı

    - İlk kısımda gördüğümüz gibi bir önceki hücreden (geçmişten) gelen bilgiyle (short term memory) şu anki girdiyi sigmoidten geçirip çarpma işlemi uygulayarak bu bilginin uzun süreli hafızaya eklenip eklenmeyeceğine karar veriyoruz.
    - İkinci kısımda, bir önceki hücreden (short term memory) gelen bilgiyi, şu anki girdiyi alıyoruz. Sigmoid ve hiperbolik tanjant fonksiyonuna sokuyoruz. Çarpma ve toplama (yani uzun süreli belleğe bilgiyi ekleme)  işlemi yapıyoruz. Buradaki çarpma ve toplama da hücre durumunun güncellenmesi işlevini de temsil ediyor.
    - Üçüncü ve son kısımda, güncellenmiş hücre durumu üzerinden gelen veriyi hiperbolik tanjant, şu anki veriyi sigmod fonksiyonuna sokarız. Bu ikisini çarparak şu anki çıktıyı elde ederiz.
