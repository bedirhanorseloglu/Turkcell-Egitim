## GPT İle Basit Metin Üretimi

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# Modeli ve tokenizer'ı yükleyelim

model_name = 'gpt2'  # Küçük GPT-2 modeli
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
# Örnek başlangıç metni

text = "I go to swim for"
```

```python
# Örnek metnin tokenizasyonu

tokenized_text = tokenizer.encode(text, return_tensors="pt") # return_tensors="pt" ifadesi, token'ların PyTorch tensörleri olarak döndürülmesini sağlar. "Tensör", çok boyutlu dizileri temsil eden veri yapılarıdır ve PyTorch gibi derin öğrenme kütüphanelerinde yaygın olarak kullanılır. Bu, modelin token'ları doğrudan işleyebilmesi için gereklidir.
```

```python
# Metin üretimi

output = model.generate(tokenized_text, max_length=10)
output
```

```python
# Şu an metin üretimi yapıldı ancak token yani sayılar halinde çıktı alındı. Bunu tekrar metne yani string'e çevirelim.
# Decode işlemini gerçekleştirelim

generated_text = tokenizer.decode(output[0], skip_special_tokens=True) # skip_special_tokens=True ifadesi, modelin ürettiği özel token'ların (örneğin, başlangıç ve bitiş token'ları gibi) metne dahil edilmemesini sağlar. Bu, çıktının daha okunabilir ve anlamlı olmasını sağlar.

generated_text
```
