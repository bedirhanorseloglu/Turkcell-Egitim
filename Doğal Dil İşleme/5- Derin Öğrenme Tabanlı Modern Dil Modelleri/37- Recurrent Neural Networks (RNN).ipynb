1. Zaman Serisi ve Sekans (Sıralı) Verisi
    - Zaman Serisi Verisi, zaman içinde ardışık olarak kaydedilen verilerdir.
        - Finans verisi (hisse fiyatları)
        - Ses kaydı verisi
        - Hava durumu verisi (havanın sıcaklığının yarın ne olacağı önceki günlere bakılarak tahmin edilir)

    - Sekan (Sıralı) Verisi, sıralı bağımlılıkları olan veridir.
        - Doğal dil
        - DNA dizileri
    
    - Dil ve Zaman Serisi Verilerinde Dizisel Bağımlılık Nedir?
        - Dizisel Bağımlılık: Zaman serisi ve dil verilerinde, her bir öğe (kelime veya veri noktası) sırasıyla önceki öğelere bağımlıdır. Cümle oluşturmak için kelimelerin belirli bir sırayla gelmesi diyebiliriz.

        - Cümlede bir kelimenin anlamı önceki kelimelerle ilişkili olabilir. Örneğin, "Ben kahve içiyorum" cümlesindeki "içiyorum" kelimesinin anlamı önceki kelimelerden etkilenir.


2. Derin Sinir Ağları Zaman Serilerinde Neden Yetersiz
    - Derin sinir ağlarının 2 temel eksikliği vardır ve bunları gidermek için RNN çıkmıştır:
        - Girdi ve çıktı boyutunun sabit olması (RNN 1 to 1 (görüntüye metin oluşturma), 1 to many (video sınıflandırma) ve many to many (çeviri) olarak farklı mimarilere ayrılır)
        - Bir hafızasının olmaması (RNN kendi çıktısını kendisine girdi olarak vererek bir hafıza sağlar ve bağımlılıkları yakalar.)
        



3. RNN Nedir
    - RNN'ler sekans (sıralı) verileri işlemek için özel olarak tasarlanmış sinir ağlarıdır
    - Her zaman adımında, önceki zaman adımındaki bilgiyi saklayarak ve sonraki adımlarda bu bilgiyi güncelleyerek çalışırlar

5. Vanishing Gradient (Kaybolan Gradyan Problemi)
    - RNN'lerde eğitim sırasında ortaya çıkan bir problemdir
    - Öğrenmenin durması demektir (ağırlıkların güncellenmemesi)
    - Backpropagation sırasında gradyanlar çok küçük hale gelir ve bu, uzun süreli bağımlılıkların öğrenilmesini zorlaştırır
    - Neden Oluşur?:
        - Her zaman adımında zincirleme türev alındığı için derin ağlarda bu türev zaman içinde küçülebilir ve neredeyse sıfıra yaklaşır
        - Bu durumda, önceki adımlardaki bilginin etkisi kaybolur.
        - Sonuç olarak kısa dönem bağımlılıkları öğrenir, ancak uzun dönem bağımlılıkları öğrenmekte zorlanır
        - Bunu gidermek için de yani bilgiyi ağın daha derinlerine iletmek için **LSTM** mimarisi geliştirilmiştir.
