# LLaMA ile Metin Üretimi

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
# Model ve tokenizer'ı yükleyelim

model_name ="huggyllama/llama-7b" # Hugging Face üzerinde bulunan LLaMA modelinin adı. 7b = 7 milyar parametreli
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Örnek başlangıç metni belirleyelim

text = "I go to swim for"
```

```python
# Tokenizasyon işlemi yapalım

tokenized_text = tokenizer.encode(text, return_tensors="pt")
```

```python
# Metin üretimi yapalım

output = model.generate(tokenized_text, max_length=15)
output
```

```python
# Şu an metin üretimi yapıldı ancak token yani sayılar halinde çıktı alındı. Bunu tekrar metne yani string'e çevirelim.
# Decode işlemini gerçekleştirelim

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
generated_text
```
