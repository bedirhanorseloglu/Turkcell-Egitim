## Word Embeddings Yöntemleri:

1. Word2Vec Nedir?
    - Anlamsal yakınlıkları kelime komşuluklarına göre öğrenir.
    - Kelimeler arasında anlamlı matematiksel işlemler yapmayı sağlar.
    - Google, Tomas Mikalov ve ekibi tarafından geliştirilmiştir
    - Google News Corpus veri seti kullanılmıştır
        - Haber metinleri
        - 100 milyar kelime
        - Siyaset, ekonomi, spor, teknoloji vb.
        - 3 milyonun üzerinde farklı kelime ve kelime çifti (zengin bir kelime ağı)

    - İki temel yöntemi vardır:
        - **CBOW (Continuous Bag of Words)** -> kelimenin bağlamındaki diğer kelimelere bakarak hedef kelimeyi tahmin etmeye çalışır. Daha hızlı öğrenme sağlar. Küçük veri setlerinde etkili çalışır.
            - İlk olarak kelimenin etrafındaki bağlam kelimeleri (örneğin bir cümledeki diğer kelimeler) tespit edilir
            - Çıktı olarak ise bağlam kelimelerine bakarak hedef kelime tahmin edilir. 
        - **Skip-gram** -> CBOW'un tersine çalışır. Hedef kelimeyi kullanarak bağlamındaki kelimeleri tahmin etmeye çalışır. Daha büyük veri setlerinde başarılıdır.

2. FastText Nedir?
    - Kelimeleri karakter seviyesinde de öğrenir.
    - Nadir kelimeler için daha güçlü bir modeldir.
    - Meta tarafından geliştirilmiştir
    - Wikipedia, Common Crawl gibi büyük veri setleri kullanılmıştır.

    


- Vektör uzayı ve anlam yakınlığı nedir?
    - Kelimelerin vektör temsilleri, anlamlarına göre vektör uzayında konumlandırılır.
    - Anlamca benzer kelimeler (kedi-köpek) vektör uzayında birbirine yakın olurken, farklı anlamdaki kelimeler (kedi-araba) uzakta yer alır
    - Yakınlık ve uzaklık ayrımı Kosinüs Benzerliği ile yapılabilir
        - İki kelimenin vekörleri arasındaki açıyı ölçen bir benzerlik metriğidir.
        - -1 ile 1 arasında bir değer alır
            - 1'e yakın değerler kelimelerin birbirine anlamca yakın olduğunu
            - 0'e yakın değerler kelimelerin birbirine anlamca farklı olduğunu belirtir
            - -1'e yakınsa bu iki kelimenin birbirine zıt olduğunu gösterir.

            


- Benzerliklerin görselleştirilmesi
