```python
import pandas as pd
import spacy 
```

```python
# Spcay kütüphanesi içerisinde bulunan bir dil modeli kullanacağız. Onu yükleyelim.

nlp = spacy.load("en_core_web_sm") # bu model İngilizce dil modeli. Türkçe dil modeli de var ama o henüz tam olarak gelişmediği için biz İngilizce modelini kullanacağız.
```

```python
# Veri seti için kendimiz bir cümle oluşturalım

content = "John works at Microsoft and lives in New York. He visited the National Museum of Natural History last week."
```

```python
# Bu content içerisinde yazdığımız cümledeki varlıkları tespit etmeye çalışacağız.
# Bunun için dil modelimize bu cümleyi işlemesi için vereceğiz.

document = nlp(content)
```

```python
#For döngüsü ile document içerisindeki her bir varlığı tek tek dolaşarak varlık ismi ve türünü yazdıralım.
for ent in document.ents:
    print(ent.text, ent.label_)


# ent.text: tokenin metin değeri (string)
    # ent.label_: underscore ile - POS etiketini OKUNABİLİR şekilde döner (örn: "NOUN", "VERB")
    # (underscore olmayan token.pos ise hash/integer ID döner - daha hızlı ama anlaşılmaz)
```

```python
# Bunu biraz daha güzel göstermek için bir tabloya dönüştürelim.

entities = [(ent.text, ent.label_) for ent in document.ents]
entities_df = pd.DataFrame(entities, columns=["Entity", "Type"])
entities_df
```
