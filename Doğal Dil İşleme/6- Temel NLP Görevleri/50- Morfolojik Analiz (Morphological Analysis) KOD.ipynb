```python
import spacy
```

```python
# Spacy kütüphanesinden bir dil modeli kullanacığımız için onu yükleyelim.

nlp = spacy.load("en_core_web_sm")
```

```python
# İncelemek için örnek bir cümle yazalım

word = "I have been running in the park for 3 hours."
```

```python
# Kelimeyi dil modelimize işlemesi için verelim.

doc = nlp(word)
doc
```

```python
# Dizi şeklinde bir kelime listesi verseydik:

# word = ["books", "RUNNING", "Happily"]
# doc = nlp(" ".join(word))
# doc
```

```python
# for dönügüsü ile tüm kelimler için morfolojik analiz yapalım.

for token in doc:
    print(f"Text: {token.text}")
    print(f"LEMMA: {token.lemma_}")
    print(f"POS: {token.pos_}")
    print(f"TAG: {token.tag_}") # ayrıntılı pos etiketi
    print(f"DEP: {token.dep_}") # cümledeki sözcük ilişkisi
    print(f"SHAPE: {token.shape_}") # kelimenin şekli (büyük harf, küçük harf, rakam vs.). 4 tane x kelime 4 küçük harften oluşuyor demektir.
    print(f"IS_ALPHA: {token.is_alpha}") # kelime sadece harflerden mi oluşuyor? Alfabetik mi?
    print(f"IS_STOP: {token.is_stop}") # kelime stop word mü?
    print(f"Morphology: {token.morph}")
    print(f"Is plural: {'Number=Plur' in token.morph}") # kelimenin çoğul olup olmadığı
    print("-----------")
```
