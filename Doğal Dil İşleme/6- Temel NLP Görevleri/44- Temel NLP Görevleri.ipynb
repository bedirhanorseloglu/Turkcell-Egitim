1- Metin Sınıflandırma (Text Classification): Elimizde bulunan ham verinin konu başlıklarına göre sınıflandırılması. Bu derste makine öğrenmesi modeli kullanarak bir çözüm geliştireceğiz.

2- Varlık İsmi Tanıma (Named Entity Recognition): Bir metin içerisinde geçen kişi, bölge, zaman gibi önemli kavramları tespit etmesi. Bu derste spacy dil modeli kullanarak bir çözüm geliştireceğiz.

3- Morfolojik Analiz (Morphological Analysis): Kelimelerin kök formlarını, çekimlerini, türetme işlemlerini ve dilbilgisel açıdan temel özelliklerini (isim, fiil, sıfat vb.) belirlemesi.  Bu derste spacy dil modeli kullanarak bir çözüm geliştireceğiz.

4- Metin Parçası Etiketleme (Part Of Speech Tagging): Bir cümledeki her kelimenin dilbilgisel kategorisini (isim, fiil, sıfat, zarf, ortaç vb.) belirlemesi. Bu derste spacy dil modeli kullanarak bir çözüm geliştireceğiz.

5- Kelime Anlamı Belirsizliği Giderme (Word Sense Disambiguation): Birden fazla anlamı olan kelimelerin cümle bağlamına göre doğru anlamının belirlenmesi. Bu derste Lesk algoritması kullanarak bir çözüm geliştireceğiz. 

6- Duygu Analizi (Sentiment Analysis): Bir metindeki olumlu, olumsuz veya nötr duygusal ifadeleri tespit edip sınıflandırması. Bu derste makine öğrenmesi modelleri ve önceden eğitilmiş modeller kullanarak bir çözüm geliştireceğiz.
