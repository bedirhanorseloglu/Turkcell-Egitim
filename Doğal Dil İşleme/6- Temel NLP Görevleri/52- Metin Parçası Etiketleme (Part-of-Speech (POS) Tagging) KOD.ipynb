#### Aslında zaten Morfolojik Analiz kısmında bunu bir nevi yapmıştık

```python
import spacy
```

```python
# Dil modelini yükleyelim

nlp = spacy.load("en_core_web_sm")
```

```python
# Örnek cümlemizi yazalım

# sentence = "What is the weather like today"

sentence = "I went to the store but they were closed, so I had to go another store"
```

```python
# Dil modelimizi cümlemize işlemesi için verelim.

doc = nlp(sentence)
doc
```

```python
# Bu tokenlarımızın her birinin hangi türde olduğunu görelim

for token in doc:
    # token.text: tokenin metin değeri (string)
    # token.pos_: underscore ile - POS etiketini OKUNABİLİR şekilde döner (örn: "NOUN", "VERB")
    # (underscore olmayan token.pos ise hash/integer ID döner - daha hızlı ama anlaşılmaz)
    print(token.text, token.pos_)
```

```python
# Bunu daha güzel gözükmesi için tablolaştıralım

import pandas as pd

doc_df = [(token.text, token.pos_) for token in doc]

table = pd.DataFrame(doc_df , columns=['text', 'pos'])
table
```
