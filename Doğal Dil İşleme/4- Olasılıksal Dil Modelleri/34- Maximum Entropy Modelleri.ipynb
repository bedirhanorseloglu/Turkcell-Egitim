- Maximum Entropy Modelleri (MaxEnt), bir olasılık dağılımını tahmin ederken mümkün olduğunca az varsayımda bulunmayı hedefler.

- Varsayımda bulunmak yerine bilinen bilgileri tam anlamıyla kullanarak olasılıkları tahmin etmeye çalışır. Olasılıkları tahmin ettikçe de zaten bir cümleyi ya da bir metni bu olasılıklar doğrultusunda oluşturmayı hedefler.

- Maksimum Entropy ilkesi, bilinmeyen bir olasılık dağılımını tahmin ederken mümkün olduğunca düzensiz (entropisi en yüksek) dağılımı seçmeyi öngörür. Bunun anlamı, eldeki verilere dayanarak hiçbir gereksiz varsayım yapılmaması gerektiğidir.

- Entropy, bir dağılımın düzensizliğinin ölçüsüdür. Maksimum Entropy kavramı da en az önyargıya sahip dağılımı temsil etmek için kullanılmaktadır.



- Kullanım Alanları
    - Özellikle sınıflandırma görevlerinde kullanılır (Örneğin bir cümlenin belirli bir sınıfa (pozitif, negatif) ait olma olasılığını tahmin edebilir)

- Avantajlar
    - Esneklik (öznitelikler arasında herhangi bir bağımsızlık varsayımı gerektirmez. Öznitelikler (kelimelerimiz) biribiri arasında bağımlı olabilir)
    - İyi genelleme (eğitim verilerine aşırı uymaktan kaçınarak yeni veriler üzerinde genelleme gerçekleştirebilir)

- Dezavantajlar
    - Hesaplama maliyeti (eğitim süreci özellikle büyük veri kümelerinde maliyetli olabilir)
    - Özellik mühendisliği (çoğunlukla doğru ve anlamlı özniteliklerin seçilmesine bağlıdır. Seçilemeze bu modellerle geliştirilen modellerimiz başarısız olur)
