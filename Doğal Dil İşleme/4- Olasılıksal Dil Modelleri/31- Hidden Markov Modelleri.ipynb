- Gizli Markov Modelleri, bir dizi gözlemin (kelimeler veya karakterler gibi) arkasında gizli bir durum dizisinin (örneğin dil bilgisel kategoriler) olduğu varsayımına dayanır.

- HMM, her bir durumun belirli bir olasılıkla başka bir duruma geçeceğini ve her durumun belirli bir gözlemi üreteceğini varsayar.



- Kullanım Alanları:
    - Konuşma tanıma
    - Dil modelleme
    - Parça etiketleme (bir cümledeki kelimelere dil bilgisel etiket atama (özne-fiil ....))

- Avantajlar:
    - Bağlam modelleme (durumlar arasındaki geçişleri ve durumdan gözleme olan olasılıkları modelleyebilir)
    - Verimli bir algoritma

- Dezavantajlar:
    - Basitleştirici varsayımlar (bağımlılıkların sadece bir önceki durumdan geldiğini varsayar. Bu da bazı durumlarda gerçekçi olmayabilir)
    - Eğitim zoruluğu (yeterli veri bulunamadığında geçiş olasılıkları doğru tahmin edilemeyebilir.)

"Gizli" (Hidden) kelimesi kullanılıyor çünkü:

Modelde iki katman vardır:

    1. Gözlenen (Observable) Katman - Biz bunları görebiliriz
        Kelimeler, sesler, karakterler gibi çıktılar

    2. Gizli (Hidden) Katman - Biz bunları doğrudan göremeyiz

        Bu gözlemleri üreten gerçek durum dizisi
        Örneğin: POS etiketleri, akustik özellikler, vb.

Somut Örnek:

- Konuşma tanımada:
    - Gözlenen: Dinlediğimiz ses sinyali
    - Gizli: Sesin hangi kelimeyi temsil ettiği (biz ses dalgasını görürüz ama arkasındaki kelime bilgisini görmeyiz)

- POS Tagging'de:
    - Gözlenen: "Araba hızlı gidiyor" cümlesinde "araba", "hızlı", "gidiyor" kelimeleri
    - Gizli: Her kelimenin dilbilgisel kategorisi (İsim, Sıfat, Fiil)

Yani model, görünür olan gözlemlerden, arkasındaki gizli durumları tahmin etmeye çalışır. İşte bu yüzden "Gizli" Markov Modeli adı verilir.
