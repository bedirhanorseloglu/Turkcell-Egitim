```python
import nltk
from nltk.tag import hmm
from nltk.corpus import conll2000
```

```python
# Gerekli veriyi yükleyelim

nltk.download("conll2000")
```

```python
# conll veri setini yükleyelim

train_data = conll2000.tagged_sents("train.txt")
test_data = conll2000.tagged_sents("test.txt")

train_data[0] # veri setinden örnek bir cümle ve etiketleri
```

```python
# Hmm eğitimi

trainer = hmm.HiddenMarkovModelTrainer()

hmm_tagger = trainer.train(train_data)
```

```python
 # Test edelim

test_sentence = "I am not going to park".split() # liste haline getiriyoruz split ile

tags = hmm_tagger.tag(test_sentence)

print(f"Test cümlesi: {tags}")
```
