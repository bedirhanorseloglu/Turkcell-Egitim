- Dil modeli, bir dildeki kelimelerin hangi sırayla geldiğini tahmin etmeye çalışan bir yapay zeka modelidir. Örneğin:
    - “Ben bugün okula ___.” → Model “gittim” gibi olası kelimeleri tahmin eder.

- Dil Modeli Nasıl Oluşturulur?
    - Veri Toplama: Büyük miktarda metin verisi (kitaplar, makaleler, sosyal medya, vs.) toplanır.
    - Önişleme: Veriler temizlenir, tokenize edilir (kelimelere bölünür), durak kelimeler çıkarılır.
    - Model Eğitimi: Makine öğrenimi veya derin öğrenme algoritmalarıyla model eğitilir.
    - Örnek modeller: n-gram, RNN, Transformer, BERT, GPT
    - Değerlendirme: Modelin doğruluğu ve anlamlılığı test edilir.
    - Kullanım: Sohbet botları, otomatik çeviri, duygu analizi gibi uygulamalarda kullanılır.

- Amaçları Nelerdir?
    - Metin üretmek (örneğin: ChatGPT gibi)
    - Cümle tamamlama
    - Soru yanıtlama
    - Duygu analizi
    - Otomatik özetleme


- Olasılıklsal dil modelleri, dilin yapısını ve düzenini anlamak ve modellemek için kullanılan istatistiksel yöntemlerdir.
- Bu modeller, bir dildeki kelimelerin dizilim olasılıklarını tahmin ederler ve genellikle metin üretimi, otomatik çeviri, dil tanıma, metin sınıflandırma gibi farklı nlp görevlerinde kullanılırlar.

    - Gizli Markov Modelleri (Hidden Markov Models - HMM)
    - Maximum Entropy Modelleri (MaxEnt)
