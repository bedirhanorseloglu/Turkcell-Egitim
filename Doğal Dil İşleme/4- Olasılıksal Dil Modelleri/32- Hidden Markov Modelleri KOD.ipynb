```python
import nltk
from nltk.tag import hmm
```

```python
# Örnek veri seti

training_data = [
    [("I", "PRP"), ("am", "VBP"), ("a", "DT"), ("student", "NN")],
    [("You", "PRP"), ("are", "VBP"), ("a", "DT"), ("teacher", "NN")],
]

# HMM'de br varsayım belirlememiz lazım başta. Örneğin I için "PRP" (zamir), am için "VBP" (fiil), a için "DT" (belirteç) ve student için "NN" (isim) gibi etiketler kullanabiliriz.
```

```python
# HMM eğitimi

trainer = hmm.HiddenMarkovModelTrainer()
hmm_tagger = trainer.train(training_data)
```

```python
# Yeni cümle ile test edelim

test_sentences = "I am a teacher".split()

tags = hmm_tagger.tag(test_sentences)

print(f"Etiketli cümle: {tags}")
```
