- Aslında N-Gram modellerinin ne olduğunu "Metin Temisli" kısmında öğrenmiştik. Ama orada metin temsili için kullanmıştık. Burada ise N-Gram modelleri kullanarak bir dil modeli oluşturmayı öğreneceğiz.
- N-Gram modelleri, bir dizideki (genellikle bir cümledeki) ardışık kelime veya karakter gruplarının olasılıklarını tahmin eder. Burada "N" dediğimiz kavram bu gruptaki öğe sayısını belirtir.



- N-Gram Modelleri:
    - Kullanım:
        - N-Gramlar; bir cümlede, bir kelimenin hangi olasılıkla diğer bir kelimeden sonra geleceğini tahmin etmek için kullanılır
    - Avantajlar:
        - Basit ve hızlı (hesaplama açısından)
        - Yerel bağıntıları iyi yakalar (ör. "ben" kelimesinden sonra "geldim" kelimesinin gelme olasılığının yüksek olmasını gayet iyi öğrenir)
    - Dezavantajları:
        - Bağlam sınırlaması (her ne kadar yerel (local) bağlantıları iyi öğrenmiş olsa da uzun mesafeli bağlamlıkları yakalayamaz)
        - Veri gereksinimi (yüksek "N" değerleri daha büyük veri kümeleri gerektirmektedir. Aksi halde buradaki "N" değerini yüksek seçemezsek nadir görülen "N" gramlar için düşük frekans sorunu yaşanacaktır.)
