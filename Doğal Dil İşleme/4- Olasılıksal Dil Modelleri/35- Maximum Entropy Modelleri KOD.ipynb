- Cümlenin duygusunu tahmin edeceğiz.

```python
from nltk.classify import MaxentClassifier
```

```python
# Eğitim veri seti

train_data = [
    ({"love":True , "amazing":True}, "positive"),
    ({"hate":True , "terrible":True}, "negative"),
    ({"happy":True , "joy":True}, "positive"),
    ({"sad":True , "depressed":True}, "negative")
]

# "love":True -> love kelimesinin kullanıldığı anlamına geliyor.
```

```python
# MaxEntropy sınıflandırıcısını eğitelim

classifer = MaxentClassifier.train(train_data)

# Default olarak 100 iterasyon yaptı.
# Bu sonuçlar bize şunu veriyor:
    # Bu veriler bize modelin her iterasyonda nasıl geliştiğini gösteriyor. 
    # Log Likelihood değeri azaldıkça modelin veriye daha iyi uyum sağladığını görebiliriz. 
    # Accuracy değeri ise modelin doğruluğunu gösterir ve bu örnekte 100 iterasyon sonunda %100 doğruluk elde edilmiştir.
```

```python
# Test Edelim

test_sentence1 = "I love this amazing movie"

# Bu cümleyi test edebilmek için elimizde bulunan kelimelerle test cümlemizdeki kelimelerin eşleşip eşleşmediğine bakmamız gerekiyor. 
# Yani test cümlemizde "love" ve "amazing" kelimeleri var, bu yüzden bu kelimelerin True olduğu bir özellik sözlüğü oluşturalım.
features = {word: (word in test_sentence1.lower().split()) for word in ["love", "amazing", "hate", "terrible", "happy", "joy", "sad", "depressed"]} 

label = classifer.classify(features)
print(f"Cümle: '{test_sentence1}' -> Tahmin Edilen Etiket: {label}")
```

```python
features 

# İlk test cümlemiz : I love this amazing movie idi. 
# Benim eğitim verimdeki "love" ve "amazing" kelimeleri bu cümlede de bulunduğu için "True" değerini aldı. 
# Bunlar da positive olarak etiketlendiğpi için model bu cümleyi "positive" olarak sınıflandırdı.
```

```python
# İklinci test cümlemiz

test_sentence2 = "This movie is terrible and I hate it"
features2 = {word: (word in test_sentence2.lower().split()) for word in ["love", "amazing", "hate", "terrible", "happy", "joy", "sad", "depressed"]}

label2 = classifer.classify(features2)
print(f"Cümle: '{test_sentence2}' -> Tahmin Edilen Etiket: {label2}")
```

```python
# Üçüncü test cümlemiz

test_sentence3 = "I do not like this movie" # "like" kelimesi eğitim verimizde yok örneğin

features3 = {word: (word in test_sentence3.lower().split()) for word in ["love", "amazing", "hate", "terrible", "happy", "joy", "sad", "depressed"]}

label3 = classifer.classify(features3)

print(f"Cümle: '{test_sentence3}' -> Tahmin Edilen Etiket: {label3}")

# ELimizde kısıtlı eğitim verisi olduğu için yanlış sınıflandırma yapabilir.
```
