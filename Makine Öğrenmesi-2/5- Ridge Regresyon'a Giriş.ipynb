Temel amacımız hata kareler toplamını minimize eden katsayıları, bu katsayılara bir ceza uygulayarak bulmak.



# RİDGE REGRESYONUNUN ÖZELLİKLERİ

- Aşırı öğrenmeye karşı dirençlidir.


- Yanlıdır fakat varyansı düşüktür. (Bazen yanlı modelleri daha çok tercih ederiz)

- Çok fazla parametre olduğunda EEK'ye (klasik regresyon) göre daha iyidir.

- Çok boyutluluk lanetine (değişken sayısının gözlem sayısından fazla olaması ) karşı çözüm sunar.

- Çoklu doğrusal bağlantı problemi (bağımsız değişkenler arasında yüksek korelasyon olması ; bir değişkenin taşıdığı bilginin aynısını diğer değişkenin de taşıması) olduğunda etkilidir.

- Tüm değişkenler ile model kurar. İlgisiz değişkenleri modelden çıkarmaz, katsayılarını sıfıra yaklaştırır.

- λ kritik roldedir. İki terimin (formüldeki) göreceli etiklerini kontrol etmeyi sağlar.

- λ için iyi bir değer bulunması önemlidir. Bunun için CV yöntemi (cross validation - çapraz doğrulama) kullanılır.

# λ AYAR PARAMETRESİNİN BELİRLENMESİ

- λ kullanıcı tarafından verilir, ayarlanır.

- β'lar bulunacak parametreler (veri setinden bulunur.)

- λ'ya sıfır verdiğimizde sonuç EEK'dir (klasik regresyon). HKT'yi (hata kareler toplamı ; SSE) minimum yapan λ'yı arıyoruz.

- λ için belirli değerleri içeren bir küme seçilir ve her birisi için cross validation test hatası hesaplanır.

- En küçük cross validation'ı veren λ, ayar parametresi olarak seçilir.

- Son olarak seçilen bu λ ile model yeniden tüm gözlemlere fit edilir.
